```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mob_df = pd.read_csv("/Users/martloni/AMR_Transmission/E_faecalis_analysis/E_faecalis_mobsuite_report.txt", sep='\t', 
                    names = ['plasmid_id', 'num_contigs', 'total_length', 'predicted_mobility', 'NCBI-HR-rank', 'NCBI-HR-Name'])
```

```python
mob_df['plasmid_id'].unique()
```

```python
mob_df.head()
```

```python
mob_df['plasmid_id'].value_counts()
```

```python
plt.style.use('ggplot')
plt.rcParams['axes.labelweight'] = 'bold'
mob_df['plasmid_id'].value_counts().plot(kind = 'bar', figsize=(20,12), fontsize = 20)
plt.title(r'Predicted plasmids from $Enterococcus$ $faecalis$ assemblies', fontsize = 30, fontweight = 'bold') #use r in front and dollar sign to make text italic
plt.xlabel('Plasmid_id', fontsize = 30)
plt.ylabel('Count', fontsize = 30)
plt.show()
plt.savefig('/Users/martloni/AMR_Transmission/E_faecalis_analysis/E_faecalis_plasmid_analysis.png', 
            dpi = 600, bbox = 'tight')
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 10))
a = sns.countplot(x=mob_df['predicted_mobility'])
a.set_xlabel("Mobility", fontsize = 20)
a.tick_params(labelsize=20)
a.set_ylabel("Count", fontsize = 20)
a.set_title(r"Plasmid mobility prediction from $Enterococcus$ $faecalis$ assemblies", fontsize = 20)
plt.ylim(0,530)
plt.tight_layout()
```

```python
mob_df['NCBI-HR-Name'].value_counts().plot(kind = 'bar', figsize=(20,12), fontsize = 20) 
plt.title(r'MOB-suite taxonomic name predictions from $Enterococcus$ $faecalis$ assemblies', fontsize = 30, fontweight = 'bold') #use r in front and dollar sign to make text italic
plt.xlabel('Predicted taxonomic name', fontsize = 30)
plt.ylabel('Count', fontsize = 30)
plt.show()
plt.savefig('/Users/martloni/AMR_Transmission/E_faecalis_analysis/E_faecalis_taxonomic_prediction.png', 
            dpi = 600, bbox = 'tight')

```
