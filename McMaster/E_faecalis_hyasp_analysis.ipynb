## *E_faecalis* hyasp plasmid analysis

#### The purpose of this notebook is to show the analysis of the E_faecalis genome assemblies with HyAsp 

* Total number of fasta file: 365
* Total number of predicted plasmid seeds: 1881

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hyasp_df = pd.read_csv("/Users/martloni/AMR_Transmission/E_faecalis_analysis/E_faecalis_hyasp_result.txt", delimiter=",",
                      names= ['Plasmid_seed_number', 'Plasmid_circularity', 'num_cds'])
```

```python
hyasp_df.head()
```

```python
hyasp_df.shape
```

```python
hyasp_df['Plasmid_seed_number'].nunique() # Gives total number of unique seeds present
```

```python
hyasp_subset = hyasp_df['Plasmid_seed_number'].value_counts().head(30) #top 30 plasmid seeds. 
```

```python
import seaborn as sns
```

```python
plt.style.use("bmh")
plt.rcParams['axes.labelweight'] = 'bold'
hyasp_subset.plot(kind = 'bar', figsize=(20,12), fontsize = 20 )
plt.title(r'Hyasp plasmid prediction from $Enterococcus$ $faecalis$ assemblies', fontsize = 30, fontweight = 'bold') #use r in front and dollar sign to make text italic
plt.xlabel('Plasmid_seed', fontsize = 30)
plt.ylabel('Count', fontsize = 30)
plt.ylim(0,50)
plt.tight_layout()
plt.show()
plt.savefig('/Users/martloni/AMR_Transmission/E_faecalis_analysis/E_faecalis_hyasp_plasmid_seed.png', 
            dpi = 600, bbox = 'tight')
```

#### The figure above is showing the top 30 putative plasmids seeds as predicted from the *Enterococcus faecalis* assemblies. 

```python
hyasp_df['Plasmid_circularity'].value_counts().sort_values()
```

```python
plt.style.use('seaborn-bright')
hyasp_df['Plasmid_circularity'].value_counts().sort_values(ascending = True).plot(kind = 'barh')
plt.title(r'Predicted plasmid circularity from $Enterococcus$ $faecalis$ assemblies', fontweight = 'bold') #use r in front and dollar sign to make text italic
plt.xlabel('Count')
plt.ylabel('Plasmid_circularity')
plt.show()
plt.savefig('/Users/martloni/AMR_Transmission/E_faecalis_analysis/E_faecalis_hyasp_plasmid_cirularity.png', 
            dpi = 600, bbox = 'tight')


```

#### Circular=1 here means complete circular plasmids

* Majority of the predicted plasmids have circular status that is equal to 0.
